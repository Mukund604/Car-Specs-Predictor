```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
from datetime import datetime
warnings.filterwarnings('ignore')
```

```python
#Avaada windows
# df = pd.read_excel('C:\\Users\\avaada\\Documents\\GitHub\\Car-Specs-Predictor\\Filtered_Data\\Filtered_data.xlsx')
#Home windows
df = pd.read_excel('C:\\Users\\mukun\\Documents\\GitHub\\Car-Specs-Predictor\\Filtered_Data\\Filtered_data.xlsx')
df = df.dropna()
df = df[df['Acceleration (s)'] != 'n.b.']
df = df[df['Torque (system performance)'] != 'n.b.']
df['CO2 emissions (NEFZ)'] = df['CO2 emissions (NEFZ)'].str.removesuffix(suffix='g/km').astype(int)
df['Fuel type'] = np.where(df['Fuel type'] == 'Super Plus', 'Ultra premium Petrol', df['Fuel type'])
df['Fuel type'] = np.where(df['Fuel type'] == 'Normal', 'Normal Petrol', df['Fuel type'])
df['Fuel type'] = np.where(df['Fuel type'] == 'Super', 'Premium Petrol', df['Fuel type'])
df['Body type'] = np.where(df['Body type'] == 'Stufenheck', 'Sedan', df['Body type'])
df['Body type'] = np.where(df['Body type'] == 'Schrägheck', 'Hatchback', df['Body type'])
df['Body type'] = np.where(df['Body type'] == 'Kombi', 'Estate', df['Body type'])
df['Body type'] = np.where(df['Body type'] == 'Cabrio', 'Roadster', df['Body type'])
df['Torque (system performance)'] = pd.to_numeric(df['Torque (system performance)'])
df['Acceleration (s)'] = df['Acceleration (s)'].str.replace(',', '.').str.strip('s').str.strip()
df['Acceleration (s)'] = pd.to_numeric(df['Acceleration (s)'])
df['Model launch date'] = pd.to_datetime(df['Model launch date'])
df['Model end date'] = pd.to_datetime(df['Model end date'])
df['Age'] = datetime.now() - df['Model end date']
# df['Age'] = pd.to_datetime(df['Age'])
df['Age'] = df['Age'].dt.days
df['Age'] = df['Age'] / 365
df
```

```python
#Avaada windows
# output_dir = 'C:\\Users\\avaada\\Documents\\GitHub\\Car-Specs-Predictor\\Conclusions'
#Home windows
output_dir = 'C:\\Users\\mukun\\Documents\\GitHub\\Car-Specs-Predictor\\Conclusions'
def plotHist(column, col_name):
    sns.histplot(column, kde = True)
    plt.title(f'{col}')
    plt.savefig(os.path.join(output_dir, f'{col_name}.png'))
    plt.close()


cols = df.columns
numeric_cols = []
for col in cols:
    if((df[col].dtype == 'int64') | (df[col].dtype == 'float64')):
        if(col == 'Top speed (km/h)'):
            numeric_cols.append(col)
            plotHist(df[col], 'Top speed (Kmph)')
        else:
            numeric_cols.append(col)
            plotHist(df[col], col)
        
    
```

DISTRIBUTION PLOT FOR EMISSIONS FUEL WISE

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Fuel type'] == 'Diesel']['CO2 emissions (NEFZ)'], label='Diesel', kde=True)
sns.histplot(df[df['Fuel type'] == 'Premium Petrol']['CO2 emissions (NEFZ)'], label='Premium Petrol', kde=True)
sns.histplot(df[df['Fuel type'] == 'Ultra premium Petrol']['CO2 emissions (NEFZ)'], label=' Ultra Premium Petrol', kde=True)
plt.title('Distribution of CO2 Emissions for Diesel and Petrol Cars')
plt.xlabel('CO2 Emissions')
plt.legend()
plt.savefig(os.path.join(output_dir, f'Distribution of CO2 Emissions for Diesel and Petrol Cars.png'))
plt.close()
```

```python
df['Model launch date'] = df['Model launch date'].dt.year
plt.figure(figsize=(12, 6))
sns.lineplot(x='Model launch date', y='CO2 emissions (NEFZ)', data=df, marker='o')
plt.title('Year-wise Trends in CO2 Emissions')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.savefig(os.path.join(output_dir, f'Year-wise Trends in CO2 Emissions.png'))
plt.close()
```

```python
plt.figure(figsize=(8, 8))
df['Body type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Market Share by Car Type')
plt.savefig(os.path.join(output_dir, f'Market Share by Car Type.png'))
plt.close()
```

```python

correlation_matrix = df[['Power (HP)','Number of cylinders (combustion engine)','Torque (system performance)','Displacement (ccm)', 'Length (mm)','Width (mm)', 'Height (mm)','Curb weight (kg)','Wheelbase (mm)','Acceleration (s)','CO2 emissions (NEFZ)']].corr()
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.1)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, f'Correlation Heatmap.png'))
plt.close()
```
